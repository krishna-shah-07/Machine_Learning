```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import numpy as np
```

```python
# Python Program to illustrate Linear Plotting
# year contains the x-axis values and e-india & e-bangladesh are the y-axis values for plotting


year = [1972, 1982, 1992, 2002, 2012]
e_india = [100.6, 158.61, 305.54, 394.96, 724.79]
e_bangladesh = [10.5, 25.21, 58.65, 119.27, 274.87]


# Setting figure size
fig = plt.figure(figsize = (8, 6)) 
# plotting of x-axis(year) and y-axis(power consumption) with different colored labels of two countries


plt.plot(year, e_india, color ='orange',label ='India',marker ='*', markersize = 12) #marker ='o', markersize = 12, 


plt.plot(year, e_bangladesh, color ='g',label ='Bangladesh',linestyle ='dashed', linewidth = 2) # linestyle ='dashed', linewidth = 2,

 

font1 = {'family':'serif','color':'blue','size':12}
font2 = {'family':'serif','color':'darkred','size':15}
plt.grid()
# naming of x-axis and y-axis
plt.xlabel('Years',fontdict = font1)
plt.ylabel('Power consumption in kWh',fontdict = font1)

 

# naming the title of the plot
plt.title('Electricity consumption per capita of India and Bangladesh',fontdict = font2)

 

pop_a = mpatches.Patch(color='#5DC83F', label='Bangladesh')
pop_b = mpatches.Patch(color='#CE5D45', label='India')
plt.legend(handles=[pop_a,pop_b],labelspacing = 1,ncol=1) #frameon is by default true

#plt.legend()
#The attribute Loc in legend() is used to specify the location of the legend.Default value of loc is loc=best (upper left). 
#The strings upper left, upper right, lower left, lower right place the legend at the corresponding corner of the axes/figure.

 

#The attribute bbox_to_anchor=(x, y) of legend() function is used to specify the coordinates of the legend, 
#and the attribute ncol represents the number of columns that the legend has.Its default value is 1.
 

#plt.legend(["India", "Bangladesh"], loc ="lower right")
# plt.legend(bbox_to_anchor =(0.75, 1.25), ncol = 2)
# plt.fill_between(e_india, e_bangladesh, year, color='g', alpha=0.3)

plt.show()
```

```python
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])
plt.subplot(2, 2, 1) #1 row 2 columns and this in the first plot
plt.title("Plot 1")
plt.plot(x, y)

x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])
plt.subplot(2, 2, 2) #this will be in the second plot
plt.title("Plot 2")
plt.plot(x, y)

x = np.array([0, 1, 2, 3])
y = np.array([30, 40, 10, 10])
plt.subplot(2, 2, 3)
plt.title("Plot 3")
plt.plot(x, y)

x = np.array([0, 1, 2, 3])
y = np.array([3, 5, 2, 7])
plt.subplot(2, 2, 4)
plt.title("Plot 4")
plt.plot(x, y)

plt.subplots_adjust(left=0.1,
                    bottom=0.1,
                    right=0.9,
                    top=0.9,
                    wspace=0.4,
                    hspace=0.4)

plt.suptitle("SUPER PLOT")
```

```python
data = {'C':20, 'C++':15, 'JAVA':30, 'Python':35}
courses = list(data.keys())
values = list(data.values())

fig = plt.figure(figsize = (10, 5))

#bar plot
plt.bar(courses, values, color = {'maroon','blue','g','r'}, width = 0.4) #verticle chart
#plt.barh(courses, values, color = {'maroon','blue','g','r'}, height = 0.5) #horizontal chart

plt.xlabel("Courses Offered")
plt.ylabel("Number of students enrolled")
plt.title("Students enrolled in different courses")
plt.show()
```
